### 1. 선 그래프(Line plot)
- plt.plot() 함수 사용
- plt.plot([x], y, [fmt], [data]) 함수를 사용하여 x, y 축을 갖는 그래프 생성
    - x, y : array-like or scalar
    - fmt : str, optional, A format string, e.g. 'ro--' for red circles with dashed line
    - 'ro--' : fmtcolor='red', marker='o', linestyle='dashed'
    - data : indexable object, optional, DataFrame 같은 객체를 주면, x, y를 column 이름으로 줄 수 있음
- plot 설명에서 색, marker 등에 대한 정보 확인
- https://matplotlib.org/2.0.2/examples/color/named_colors.html
- https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html

#### 1-1. 하나의 선 그리기
- plt.plot(y) : x는 [0, 1, 2, ..] 자동생성
- plt.plot(x, y) : x, y 둘 다 지정, 개수가 동일해야 함
- plt.plt(x, y, fmt) : 선의 스타일 지정

````python
# plot 그리기 실습
# y 축만 지정, x 축은 [0, 1, 2, 3...] 와 같이 자동으로 생성됨 
data = np.arange(4, 14)   # [4 5 6 7 ... 13]
plt.plot(data)  # y로 사용됨
plt.grid()
plt.show()
````

````python
# Python 방식
x = range(10)
y = [d ** 2 for d in x]
print(y)
````

````python
# x는 0 ~ 9, y = x**2 에 대한 그래프 그리기
x = np.arange(10)
y = x ** 2
plt.plot(x, y)
plt.show()
````

#### 1-3. x,y 축 레이블 표시
- plt.xlabel(str, fontsize) : x 축에 대한 레이블 표시
- plt.ylabel(str, fontsize) : y 축에 대한 레이블 표시
- plt.title(str, fontsize) : plot에 대한 제목

````python
# X축  'money',  Y축 'time'  제목 : 'time for money'
x = np.arange(10)
plt.plot(x, x*2, 'b-', x, x**2, 'r--', x, x**3, 'go')
plt.xlabel('money', fontsize=12)
plt.ylabel('time', fontsize=12)
plt.title('time for money', fontsize=15)
plt.show()
````

### 6. 히스토그램 (연속형)
- 변수가 하나인 단변수 데이터의 빈도수를 그래프로 표현(일변량 그래프)
- 히스토그램을 그리기 위한 hist 명령은 bins 인수로 데이터를 집계할 구간 정보를 받음
- 데이터 집계 결과를 반환
- n, bins, patches = plt.hist(x, bins, range)
   - x : 데이터
   - bins : 계급
   - range: 대상 범위, (min, max)
   - n : 도수 
   - patches : 계급별 히스토그램 좌표와 높이
- https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html  

````python
import pandas as pd
np.random.seed(10)
x = 100 + 15 * np.random.randn(10000)
s, bins = pd.cut(x, bins=50, retbins=True)
s2 = s.value_counts()
df = pd.DataFrame(s2, columns=['count'])
df['rate'] = df['count']/ df['count'].sum()
df['density'] = df['rate']/(bins[1:]-bins[0:50])   #밀도 = 상대도수/계급구간길이
bins
````

````python
# x에 대해 50개 계급으로 나눈 히스토그램 그리기, 
# density=True 사용
np.random.seed(10)
x = 100 + 15 * np.random.randn(10000)
n, bins, patches = plt.hist(x, 50, density=True)
plt.grid()
plt.show()
````

#### 8-1. boxplot 

````python
# tips의 'total_bill' 의 분포를 boxplot으로 확인함 (x로 지정)
sns.set_style("whitegrid")
plt.figure(figsize=(8, 6))
sns.boxplot(y='total_bill', data=tips)
#sns.boxplot(x=tips['total_bill'])
plt.show()
````

````python
# 이상치(Outlier)
Q1, Q3 = tips['total_bill'].quantile([0.25, 0.75])
IQR = Q3 - Q1
lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
print(Q1, Q3, IQR, lower, upper)
````

````python
# 특정 카테고리에 대한 값의 분포는 X,Y를 지정하여 확인 가능
# x, y, data 사용
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
````

````python
# hue 옵션을 통해 카테고리 별 데이터 시각화 가능
# palette 옵션을 통해 색상 지정 가능 ('Set3' 사용)
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips)
plt.show()
````

````python
# boxplot + swarmplot
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
plt.show()
````

- 데이터 분포 및 변수간 관계를 분석

````python
sns.set(style="ticks")
iris = sns.load_dataset("iris")
iris.head(10)
````

````python
# pairplot을 이용하여 변수(컬럼)간 관계와 분포 확인 가능
#sns.set(style="ticks")
sns.pairplot(iris)
plt.show()
````

````python
# hue 옵션을 이용하여 카테고리별 변수의 관계와 분포 확인 가능
sns.pairplot(iris, hue='species')
plt.show()
````

````python
# vars 옵션을 이용하여 특정 변수만 지정 가능
sns.pairplot(iris, vars=['sepal_width', 'sepal_length'])
plt.show()
````

````python
# x_vars, y_vars 옵션을 사용하여 해당 축의 특정 변수만 지정 가능
sns.pairplot(iris, x_vars=['sepal_width', 'sepal_length'],
            y_vars=['petal_width', 'petal_length'])
plt.show()
````
